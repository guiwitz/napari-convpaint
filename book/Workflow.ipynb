# Example of a workflow

Please keep in mind that napari-convpaint is not meant to be a standalone software for segmentation. Therefore, its functionalities are limited to the pixel prediction. However, the napari ecosystem is very powerful and flexible. Here we use napari-convpaint as a first step for pixel prediction and complete the segmentation using tools bundlend in the [napari-assistant plugin](https://github.com/haesleinhuepf/napari-assistant) (developed by Robert Haase). ## Adding napari-segment-blobs-and-thing-with-membranes
After installing napari-convpaint, you can open napari, go to Plugins menu and in the "Install/Uninstall Plugins..." section you can use the filter menu to search for napari-segment-blobs-and-things-with-membranes. 

Press the install button to add it to your napari plugins.

For more information about the plugin and other ways of installing it please refer to the [docs of the plugin](https://github.com/haesleinhuepf/napari-segment-blobs-and-things-with-membranes/tree/main).
